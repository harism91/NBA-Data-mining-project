#### Drop irrelevant/repetitive variables such as NICKNAME and COMMENT

```python
penguin.drop(['NICKNAME'], axis=1, inplace=True)
penguin.drop(['COMMENT'], axis=1, inplace=True)
```

#### convert name attributes into numerical

```python
START_POSITION_num = {'F': 1, 'C': 2,  'G': 3 }
penguin['START_POSITION'] = penguin['START_POSITION'].map(START_POSITION_num)
penguin['START_POSITION'] = penguin['START_POSITION'].replace(np.nan, 4)
penguin.head(20)
```
